```python
import pandas as pd 

full_data_info = pd.read_csv('111.csv',encoding ='utf-8')
```

```python
import numpy as np
def find_outliers_3sigma(data):  
    
    if len(data)==1 or len(set(data))==1:
        one_hot_num = float(data.head(1))
        if one_hot_num<=100:
            thr_sig_num =  one_hot_num*2.5
        elif 100<one_hot_num<=500:
            thr_sig_num = one_hot_num*2
        elif 500<one_hot_num<=2000:
            thr_sig_num = one_hot_num*1.8
        else:
            thr_sig_num = one_hot_num*1.5
    else:
        thr_sig_num  = np.nan 
    if thr_sig_num<20:
        thr_sig_num = 20

    return thr_sig_num



thr_sigma_info_lower_two = full_data_info.groupby(['PARTNAME' ,'LOCNAME' ,'DEMANDSTREAMID'])['HISTORYAMOUNT'].apply(find_outliers_3sigma).reset_index()
# thr_sigma_info.sort_values(by=['HISTORYAMOUNT'])
thr_sigma_info_lower_two
```

```python
import numpy as np
def find_outliers_3sigma(data):  
    
    if len(data)==1 or len(set(data))==1:
        thr_sig_num = np.nan
    else:
        mean = sum(data) / len(data)  
        std_dev = (sum((x - mean) ** 2 for x in data) / len(data)) ** 0.5  
        thr_sig_num  = mean + 3 * std_dev 
        
    if thr_sig_num<20:
        thr_sig_num = 20

    return thr_sig_num



thr_sigma_info_3sigma = full_data_info.groupby(['PARTNAME' ,'LOCNAME' ,'DEMANDSTREAMID'])['HISTORYAMOUNT'].apply(find_outliers_3sigma).reset_index()
# thr_sigma_info.sort_values(by=['HISTORYAMOUNT'])
thr_sigma_info_3sigma
```

```python
import numpy as np
def find_outliers_3sigma(x):  
    
    if len(x)==1 or len(set(x))==1:
        thr_sig_num = np.nan
    else:
        q3 = lambda x: x.quantile(0.75)  
        q1 = lambda x: x.quantile(0.25)  
        thr_sig_num = q3(x) + 1.5 * (q3(x) - q1(x))  
    if thr_sig_num<20:
        thr_sig_num = 20

    return thr_sig_num



thr_sigma_info_xiangxingtu  = full_data_info.groupby(['PARTNAME' ,'LOCNAME' ,'DEMANDSTREAMID'])['HISTORYAMOUNT'].apply(find_outliers_3sigma).reset_index()
# thr_sigma_info.sort_values(by=['HISTORYAMOUNT'])
thr_sigma_info_xiangxingtu
```

```python
raw_data_info = full_data_info[['PARTNAME' ,'LOCNAME' ,'DEMANDSTREAMID','HISTORYBEGDATE','HISTORYAMOUNT']]
raw_data_info.head()
```

```python
# 将它们放入一个列表中  
# df_list = [raw_data_info, thr_sigma_info_lower_two,thr_sigma_info_3sigma,thr_sigma_info_xiangxingtu]  
  
# 使用functools.reduce和pandas.concat来合并所有DataFrame  
merged_df_lower_two = pd.merge(raw_data_info, thr_sigma_info_lower_two,
                              on=['PARTNAME' ,'LOCNAME' ,'DEMANDSTREAMID'],how ='left')
```

```python
merged_df_sigma = pd.merge(merged_df_lower_two, thr_sigma_info_3sigma,
                              on=['PARTNAME' ,'LOCNAME' ,'DEMANDSTREAMID'],how ='left')
merged_df_xiangxingtu = pd.merge(merged_df_sigma, thr_sigma_info_xiangxingtu,
                              on=['PARTNAME' ,'LOCNAME' ,'DEMANDSTREAMID'],how ='left')
merged_df_xiangxingtu.head()
```

```python
merged_df_xiangxingtu.columns = ['PARTNAME' ,'LOCNAME' ,'DEMANDSTREAMID',
                                 'HISTORYBEGDATE','HISTORYAMOUNT','历史记录小于1次阈值','3_sigma阈值','箱型图阈值']

merged_df_xiangxingtu.sort_values(by=['PARTNAME' ,'LOCNAME' ,'DEMANDSTREAMID',
                                 'HISTORYBEGDATE'],inplace =True)
```

```python
merged_df_xiangxingtu.columns
```

```python
merged_df_xiangxingtu.dropna(subset='PARTNAME',inplace =True)
```

```python
merged_df_xiangxingtu['历史记录小于1次拦截情况'] = np.where(merged_df_xiangxingtu['HISTORYAMOUNT']
                                                 < merged_df_xiangxingtu['历史记录小于1次阈值'], 1, 0)  
merged_df_xiangxingtu['3_sigma阈值拦截情况'] = np.where(merged_df_xiangxingtu['HISTORYAMOUNT']
                                                 < merged_df_xiangxingtu['3_sigma阈值'], 1, 0) 
merged_df_xiangxingtu['箱型图阈值拦截情况'] = np.where(merged_df_xiangxingtu['HISTORYAMOUNT']
                                                 < merged_df_xiangxingtu['箱型图阈值'], 1, 0) 
```
