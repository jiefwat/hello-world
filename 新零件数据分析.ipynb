```python
import pandas as pd 
new_2021 = pd.read_csv('202101-202106参数new.csv',encoding='gb18030')
new_2023 = pd.read_csv('202301-202303参数new.csv',encoding='gb18030')
full_data = new_2021.append(new_2023)
```

```python
def month_fillna(data):
    for i in ['1','2','3','4','5','6']:
        data[i].fillna(0,inplace =True)
    return data

full_data = month_fillna(full_data)

full_data.head()
```

```python
import numpy as np
full_data['to_2_rate'] = full_data['2']/full_data['1']
full_data['to_3_rate'] = full_data['3']/full_data['1']
full_data['to_4_rate'] = full_data['4']/full_data['1']
full_data['to_5_rate'] = full_data['5']/full_data['1']
full_data['to_6_rate'] = full_data['6']/full_data['1']
for i in ['to_2_rate','to_3_rate','to_4_rate','to_5_rate','to_6_rate']:
        full_data[i].fillna(0,inplace =True)
        full_data[i] = full_data[i].replace(np.inf,0)
full_data.head()
```

```python
high_fst_data = full_data[(full_data['1']>3)]
high_fst_data.shape
```

```python
high_nslow_data = high_fst_data[high_fst_data['ABC需求数量分类']!='SLOW']
high_nslow_data.shape
```

```python
high_nslow_data.head(1)
```

```python

test = high_nslow_data[(high_nslow_data['1']>3)&(high_nslow_data['1']<20)]

# for i in ['1','2','3','4','5']:
#         test[i]=['下降' if x<1 else '上升' for x in test[i]]

        
print('首月值在[4,20]，第2个月值的概率')
print(test['2'].value_counts(normalize=True).head(3))

print('首月值在[4,20]，第3个月值的概率')
print(test['3'].value_counts(normalize=True).head(3))

print('首月值在[4,20]，第4个月值的概率')
print(test['4'].value_counts(normalize=True).head(3))

print('首月值在[4,20]，第5个月值的概率')
print(test['5'].value_counts(normalize=True).head(3))

print('首月值在[4,20]，第6个月值的概率')
print(test['6'].value_counts(normalize=True).head(3))


# ABC价格分类 为 D的新零件，后续月份值下降的概率分别为83%，74%，77%，77%，75%
# ABC价格分类 为 A的新零件，后续月份值下降的概率分别为86%，69%，77%，74%，78%
# ABC价格分类 为 C的新零件，后续月份值下降的概率分别为72%，67%，69%，72%，68%
```
