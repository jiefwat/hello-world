```python
import pandas as pd 
data = pd.read_csv('18-23数据.csv',encoding='utf-8')


data_yichang = pd.read_csv('异常件.csv',encoding='utf-8')
data_yichang['计划员']=0

data = data.append(data_yichang)
data['日期'] = pd.to_datetime(data['日期'], format='%Y%m%d').dt.strftime('%Y-%m-%d')
new_df = data.groupby(['零件号','日期'])['需求数量'].sum().reset_index()
new_df['日期'] = pd.to_datetime(new_df['日期'])
```

```python
new_df.head()
```

```python
# 生成日期范围
date_range = pd.date_range(start='2016-12-12', end='2023-07-01', freq='D')
# 创建空的DataFrame，准备存储填充后的结果
filled_df = pd.DataFrame()

# 针对每个 SKU 进行填充操作
for sku, group in new_df.groupby('零件号'):
    sku_group = group.set_index('日期').reindex(date_range, fill_value=0).reset_index()
    sku_group['零件号'] = sku
    filled_df = filled_df.append(sku_group, ignore_index=True)

filled_df.head()
```

```python
key_index = data.drop_duplicates('零件号')
print(filled_df.shape)
data_day = pd.merge(filled_df,key_index.drop(['日期', '需求数量'], axis=1),on = ['零件号'],how ='left')
print(data_day.shape)
```

```python
data_day.rename(columns = {'index':'日期'},inplace =True)
data_day.head()
```

```python
# 将DataFrame按照sku_id进行分组，并按照date升序排序
data_day = data_day.sort_values(by=['零件号', '日期']).reset_index(drop=True)

# 定义一个函数来处理每个分组
def remove_zero_sales(group):
    if group['需求数量'].eq(0).all():
        return pd.DataFrame()  # 返回空DataFrame，表示整个分组都要去除
    else:
        first_nonzero_index = group['需求数量'].ne(0).idxmax()
        return group.loc[first_nonzero_index:]

# 对每个分组调用处理函数
data_day = data_day.groupby('零件号').apply(remove_zero_sales).reset_index(drop=True)
```

```python
post_processed_data  = data_day[['零件号','日期',
                           '需求数量','工厂','需求频次abc分类','价格abc分类','需求数量ABC分类','零件代码第四位','计划员']].sort_values(by =['零件号','日期'])
```

```python
import numpy as np
import warnings
warnings.filterwarnings("ignore")


def get_date_feature(df):
    """取时间特征"""
    df["日期"] = pd.to_datetime(df["日期"], format="%Y-%m-%d %H:%M:%S")
    df["dayofmonth"] = df["日期"].dt.day.astype(int)
    df["dayofyear"] = df["日期"].dt.dayofyear.astype(int)
    df["dayofweek"] = df["日期"].dt.dayofweek.astype(int)
    df["month"] = df["日期"].dt.month.astype(int)
    df["week"] = df["日期"].dt.weekofyear.astype(int)
    df["quarter"] = (df["日期"].dt.quarter).astype(int)
    return df


post_processed_data['日期'] = pd.to_datetime(post_processed_data['日期'])


post_processed_data.sort_values(['零件号', '日期'], inplace=True)
post_processed_data = post_processed_data.reset_index(drop=True)
```

```python
# 生成前12个月每个月销量的列
for i in range(1, 12):
    post_processed_data[f'LastMonthSales_{i}'] = post_processed_data.groupby('零件号')['需求数量'].shift(i)
```

```python
# 生成历史最大值的列
#     df['MaxSales'] = df.groupby('零件号')['需求'].shift(1).expanding().max()

#     生成近半年的最大值
post_processed_data['LastsixMonthsMax'] = post_processed_data.groupby('零件号')['需求数量'].rolling(window=3, min_periods=1).apply(lambda x: x[:-1].max()).reset_index(drop=True)
print('max')

post_processed_data['LastsixMonthsMin'] = post_processed_data.groupby('零件号')['需求数量'].rolling(window=3, min_periods=1).apply(lambda x: x[:-1].min()).reset_index(drop=True)
print('min')

post_processed_data['LastsixMonthsmean'] = post_processed_data.groupby('零件号')['需求数量'].rolling(window=3, min_periods=1).apply(lambda x: x[:-1].mean()).reset_index(drop=True)
print('mean')

post_processed_data['LastsixMonths85Quantile'] = post_processed_data.groupby('零件号')['需求数量'].rolling(window=3, min_periods=1).apply(lambda x: x[:-1].quantile(0.25)).reset_index(drop=True)
print('85')

post_processed_data['LastsixMonths50Quantile'] = post_processed_data.groupby('零件号')['需求数量'].rolling(window=3, min_periods=1).apply(lambda x: x[:-1].quantile(0.50)).reset_index(drop=True)
print('50')


post_processed_data = get_date_feature(post_processed_data)
post_processed_data= post_processed_data[(post_processed_data['日期']>='2017-07-01')&(post_processed_data['日期']<'2023-07-01')]
```

```python
processed_data = post_processed_data.copy()

processed_data['零件代码第四位'] = processed_data['零件代码第四位'].astype(str)

categorical_features = ['零件号','需求频次abc分类','价格abc分类','需求数量ABC分类']


processed_data[categorical_features] = processed_data[categorical_features].astype('category')
train_datasets = processed_data[processed_data['日期']<'2023-01-01']
test_datasets = processed_data[processed_data['日期']>='2023-01-01']

X_train = train_datasets.drop(['需求数量','日期'], axis=1)
y_train = train_datasets['需求数量']

X_test = test_datasets.drop(['需求数量','日期'], axis=1)
y_test = test_datasets['需求数量']


from catboost import CatBoostRegressor

# 创建CatBoost回归模型
model = CatBoostRegressor(iterations=100, learning_rate=0.1, depth=6, cat_features=categorical_features)

# 训练模型
model.fit(X_train, y_train, verbose=100)

# 进行预测
y_pred = model.predict(X_test)

# 评估模型
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
info_forecast = test_datasets.reset_index(drop=True)
info_forecast['real_values'] = pd.DataFrame(y_test).reset_index(drop=True)

info_forecast['pred_values'] = pd.DataFrame(y_pred)

info_forecast['pred_values'] = np.where(info_forecast['pred_values'] >= info_forecast['LastsixMonthsMax'], info_forecast['LastsixMonthsMax'], np.where(info_forecast['pred_values'] <= info_forecast['LastsixMonthsMin'], info_forecast['LastsixMonthsMin'], info_forecast['pred_values']))

info = info_forecast.groupby(['零件号','计划员','需求频次abc分类','month'])['real_values','pred_values'].sum().reset_index()
```

```python


# 定义区间
import matplotlib.pyplot as plt
import seaborn as sns

bins = [0, 0.10, 0.30, 0.50, 1.00, 100000]

# 使用cut函数将数据分成不同区间
info['ScoreRange'] = pd.cut(info['mape'], bins=bins, labels=['0-10%', '10%-30%', '30%-50%', '50%-100%', '>100%'],right =False)

info['ScoreRange'].replace('NaN','0',inplace =True)
# 计算每个区间中的数据数量


wmape_info  = info.groupby(['month'])['real_values','pred_values'].sum().reset_index()
wmape_info['mape'] = (abs(wmape_info['real_values'] - wmape_info['pred_values']) / wmape_info['real_values']) 


grouped = info.groupby(['month','ScoreRange']).size().reset_index(name='sku_mm_values')
# 计算每个月份内的总样本数
monthly_counts = info.groupby(['month']).size().reset_index(name='mm_values')

# 合并统计结果
result = pd.merge(grouped, monthly_counts, on=['month'],how = 'left')

# 计算占比
result['Proportion'] = result['sku_mm_values'] / result['mm_values']
```
