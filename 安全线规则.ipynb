```python
import pandas as pd 
data = pd.read_excel('库存系数_需求.xlsx')
```

```python
data.head()
```

```python
  
# 这里假设你的DataFrame已经存在并命名为df  
  
def calculate_value(row):  
    if row['是否正常采购'] == 'n':  
        return 0.00001  
    elif row['是否正常采购'] == 'y':  
        if row['是否慢流件'] in ['分中心发货', '总库集中发货']:  
            if row['是否呆滞风险'] == '呆滞风险件':  
                return 0.0001  
            elif row['上月库存结构系数'] <= 0.2:  
                return row['上月需求'] * 0.2  
            elif row['上月库存结构系数'] >= 0.3:  
                return row['上月需求'] * 0.1  
        else: # 列‘是否慢流件’的值不为‘分中心发货‘或者’总库集中发货’  
            if row['上月库存结构系数'] <= 0.2:  
                return row['上月需求'] * 0.2  
            elif row['上月库存结构系数'] > 0.2 and row['上月库存结构系数'] < 0.3:  
                return 0.0001  
            elif row['上月库存结构系数'] >= 0.3:  
                return row['上月需求'] * 0.1  
    else: # 其他情况  
        return 0.0001  
    
data['values'] = data.apply(calculate_value, axis=1)
```

```python
data.to_csv('1111.csv',encoding ='gb18030')
```
