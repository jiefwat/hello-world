```python
import pandas as pd 
data = pd.read_csv('202101-202106(1).csv',encoding='utf-8')
data['上线月份'] = pd.to_datetime(data['上线月份'], format='%Y%m').dt.strftime('%Y-%m')
data.head()
```

```python
real_demods_info = data[data['新旧']=='new']
real_demods_info = real_demods_info.fillna(1)
real_demods_info
```

```python
pred_svg_info = pd.read_csv('svg预测(2).csv',encoding='utf-8')
for i in ['1','2','3','4','5','6']:
    pred_svg_info[i] = [1 if x <0.01 else x for x in pred_svg_info[i]]
pred_svg_info = pred_svg_info.fillna(1) 
```

```python
pred_svg_info
```

```python
output = pd.merge(real_demods_info,pred_svg_info,on =['零件号','仓库'],how ='left')
output.head(10)
```

```python
output['2_mape'] = abs(output['1_y']-output['2_x'])/output['2_x']
output['3_mape'] = abs(output['2_y']-output['3_x'])/output['3_x']
output['4_mape'] = abs(output['3_y']-output['4_x'])/output['4_x']
output['5_mape'] = abs(output['4_y']-output['5_x'])/output['5_x']
output['6_mape'] = abs(output['5_y']-output['6_x'])/output['6_x']
output.head()
```

```python
mape_info  = output[['零件号','仓库','2_mape','3_mape','4_mape','5_mape','6_mape']]
mape_info.mean()
```

```python
# 计算mape，wmape 占比情况

# 定义区间
import matplotlib.pyplot as plt
import seaborn as sns

bins = [0, 0.10, 0.30, 0.50, 1.00, 100000]
test = mape_info.copy()
# 使用cut函数将数据分成不同区间
for i in ['2_mape','3_mape','4_mape','5_mape','6_mape']:
    print(test[i].mean())
    test[i] = pd.cut(test[i], bins=bins, labels=['0-10%', '10%-30%', '30%-50%', '50%-100%', '>100%'],right =False)
    test[i].replace('NaN','0',inplace =True)
# 计算每个区间中的数据数量
    print('新品第',i.split('_')[0],'月误差')
    print(test[i].value_counts(normalize=True))
```
